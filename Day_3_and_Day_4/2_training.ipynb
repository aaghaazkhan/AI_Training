Importing the libraries

```python
from ultralytics import YOLO
import torch
```

```python
torch.cuda.is_available()
```

We will be using YOLOv8s model for this task

```python
model = YOLO("yolov8s.pt")
```

```python
results = model.train(
    data="tattoo_dataset/data.yaml", # data.yaml file path
    epochs=100, # number of training epochs
    imgsz=512, # image size
    batch=32, # batch size
    optimizer="SGD", # optimizer
    lr0=0.001, # learning rate
    device=0, # device = 0 as I have only a single dedicated GPU
    name="run1", # run name
    save=True, # model saving
    save_period=10, # saving the checkpoints after every 10 epochs
    save_json=True, # save json file
    project="tattoo", # project name
    exist_ok=True, # overwriting existing project
    resume=False, # training resume
    patience=10 # early stopping value
)
```
